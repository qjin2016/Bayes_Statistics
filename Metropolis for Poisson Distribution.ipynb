```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lambda1 =  4.0
MAX_ITERATION = 100000
MAX_COUNTER = int(6 * lambda1)
n_start = 10
list_poisson = [None] * MAX_ITERATION
COUNTER = [0] * MAX_COUNTER
COUNTER_X = range(MAX_COUNTER)
COUNTER_POISSON = [0] * MAX_COUNTER
```

```python
n = n_start

for i in range(MAX_ITERATION):
    u1 = np.random.random() 
    
    if u1 > .5:
        n_trail = n + 1
    else:
        n_trail = n - 1
        
    if n_trail < 0:
        print(n)
        list_poisson[i] = n
        COUNTER[n] = COUNTER[n] + 1
        continue
        
    p_trail = math.exp(-lambda1) * (lambda1 ** n_trail) / float(math.factorial(n_trail))

    p_current = math.exp(-lambda1) * (lambda1 ** n) / float(math.factorial(n))
    
    if (p_trail > p_current):
        n = n_trail
        print(n)
        list_poisson[i] = n
        COUNTER[n] = COUNTER[n] + 1
    else:
        r = np.random.random()
        if (p_trail / p_current > r):
            n = n_trail
            print(n)
            list_poisson[i] = n
            COUNTER[n] = COUNTER[n] + 1
        else:
            print(n)
            list_poisson[i] = n
            COUNTER[n] = COUNTER[n] + 1
```

```python
for n in range(MAX_COUNTER):
    COUNTER[n] = COUNTER[n] / float(MAX_ITERATION)
    
for n in range(MAX_COUNTER):
    COUNTER_POISSON[n] = math.exp(-lambda1) * (lambda1 ** n) / float(math.factorial(n))
```

```python
plt.plot(COUNTER_X, COUNTER)

plt.plot(COUNTER_X, COUNTER_POISSON)

plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('lambda= ' + str(lambda1))

plt.show()
```

```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
np.random.random()
```

```python
def coinflip(n, m, p):
    num_heads = []
    for _ in range(m):
        num_head = 0
        for _ in range(n):
            if np.random.random() > .3:
                num_head += 1
        num_heads.append(num_head)
    sns.distplot(num_heads, rug=False, kde=False)
    print('mean: %i' % (np.mean(num_heads)))
    print('variance: %i' % (np.var(num_heads)))
```

```python
coinflip(8, 3000, .3)
```
